```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('nba2k-full.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data.describe()
data.team = data.team.fillna('Unknown')
data.college = data.college.fillna('NA')
data.salary = data.salary.apply(lambda x: x.replace('$', '')).astype(int)
df_cleaned = data.copy()
```

```python
df_cleaned = data.copy()
```

```python
data.info()
```

```python
correlation_matrix = data[["salary","rating","draft_year"]].corr()
```

```python
# Plot a heatmap for the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Show the plot
plt.title('Correlation Between NBA Player Overall Rating and Salary')
plt.show()
```

```python
plt.bar(data["position"], data["rating"])
```

```python
plt.figure(figsize=(10,6))
plt.hist(df_cleaned['salary'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of NBA Player Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df_cleaned['rating'], bins=15, color='orange', edgecolor='black')
plt.title('Distribution of NBA Player Overall Ratings')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df_cleaned['rating'], df_cleaned['salary'], color='green', alpha=0.6)
plt.title('NBA Player Salary vs. Overall Rating')
plt.xlabel('Overall Rating')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
# Select relevant columns
relevant_columns = df_cleaned[['salary', 'rating']]  # Add more columns if available

# Create the pair plot
sns.pairplot(relevant_columns)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='rating', y='salary', data=df_cleaned, palette='coolwarm')
plt.title('Violin Plot of Salary Distribution by Overall Rating Groups')
plt.xlabel('Rating Group')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
# Sort by salary and get the top 10 highest-paid players
top_10_players = df_cleaned.nlargest(10, 'salary')

plt.figure(figsize=(10,6))
sns.barplot(x='salary', y='rating', data=top_10_players, palette='Blues_d')
plt.title('Top 10 Highest-Paid NBA Players and Their Overall Ratings')
plt.xlabel('Salary')
plt.ylabel('Overall Rating')
plt.grid(True)
plt.show()

```
